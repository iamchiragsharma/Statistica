```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
label = ['C++','Python','Data Structures','Algorithms','News','Extra Knowledge']
score = [40,60,45,45,50,30]
```

```python
index = np.arange(len(label))
plt.bar(index,score)
plt.xlabel('Subject of Interest')
plt.ylabel('Current Position')
plt.xticks(index,label,fontsize=10,rotation=20)
plt.title('My performance yet')
plt.show()
```
