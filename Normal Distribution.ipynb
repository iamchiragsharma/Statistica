```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as  ss

plt.style.use('seaborn') # pretty matplotlib plots
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
heights = np.random.randint(160,high=200,size=50)
heights = np.sort(heights)
meanHeights = np.mean(heights)
varHeights = np.var(heights)
stdHeights = np.std(heights)
```

```python
print(heights)
print(meanHeights)
print(varHeights)
print(stdHeights)
```

```python
y_pdf = ss.norm.pdf(heights,meanHeights,stdHeights)
y_cdf = ss.norm.cdf(heights,meanHeights,stdHeights)
```

```python
plt.plot(heights,y_pdf)
plt.xlabel('heights')
plt.ylabel('normal distribution')
```

```python
plt.plot(heights,y_cdf)
plt.xlabel('heights')
plt.ylabel('cummulative probability')
```
