```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#Maximum Likelihood 
#Let's talk about likelihood, it means chances of something being the likely,probability so yeah they both are
#different terms but every easy to confuse with.
#Probability is obtained for a particular data with fixed parameters & likelihood is obtaining the parameters
#given a particular result and maximum likelihood is way of maximizing the probability of best observed value 
#out of the observed values.

mouse = 32
means = np.linspace(26,38,1000)
likelihood = []
for mu in means:
    likelihood.append(norm.pdf(32,mu,2))
```

```python
curve = dict(zip(means,likelihood))
```

```python
maxLikelihood = list(max(curve.items(), key=lambda k: k[1]))  
xaxis = [maxLikelihood[0],0]
yaxis = [0,maxLikelihood[1]]
```

```python
plt.plot(means,likelihood)
plt.show()
```
