```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10,6]
import seaborn as sns
import scipy.stats
```

```python
populationMarks = 100
marks = np.random.randint(1,500,populationMarks)
pMean = np.mean(marks)
pVar = np.var(marks)
print('populationMean: {} \npopulationVar: {}'.format(pMean,pVar))
```

```python
sMean = []
sVar= []
trials = 10000
sample = 20
for i in range(trials):
    sMean.append(np.mean(np.random.choice(marks,sample)))
    #If the variance is taken is sigma**2/n
    sVar.append(np.var(np.random.choice(marks,sample)))
#There is one more corrected var for the finite population.
correctedVar = (pVar/sample)*((populationMarks-sample)/(populationMarks-1))
sMean = np.array(sMean)
sVar = np.array(sVar)
print('For {} trials: \nSample Mean is {} \nSample Variance is {}'.format(trials,np.mean(sMean),np.mean(sVar)))
```

```python
plt.hist(sMean, bins=30, color='#EAC4B9', density="true")
sns.kdeplot(sMean)
plt.show()
```

```python
plt.hist(sVar, bins=30, color='#aaffc3', density="true")
sns.kdeplot(sVar)
plt.show()
```

```python
def error(first,second):
    print((first-second)/100)
```

```python
error(pMean,correctedVar)
error(pMean,(np.mean(sMean**2))**0.5)
```

```python
#Standardized Sample Mean
stdMean = (sMean-pMean)/(pVar/populationMarks)**0.5
plt.hist(stdMean, bins=30, color='#EAC4B9', density="true")
sns.kdeplot(stdMean)
plt.show()
```

```python
count,lowerlimit,dx,_= scipy.stats.cumfreq(stdMean, numbins=30)
#scipy.stats.cumfreq(stdMean, numbins=30)
```

```python
x_axis = np.arange(count.size) *dx + lowerlimit 
```

```python
plt.plot(x_axis,count, 'ro')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.show()
```
