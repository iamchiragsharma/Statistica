```python
from scipy.stats import poisson
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x= np.arange(1,10,1)
plt.plot(x, poisson.pmf(x,4.6))
plt.show()
y = poisson.pmf(x,4.6)
print(y)
```

```python
x = np.arange(1,10,1)
```

```python
plt.plot(x,poisson.cdf(x,4.6))
plt.show()
y = poisson.cdf(x,4.6)
print(y)
```
