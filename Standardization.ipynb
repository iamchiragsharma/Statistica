```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation
plt.rcParams['animation.ffmpeg_path'] = '/usr/bin/ffmpeg'
%matplotlib inline
from scipy.stats import norm
import seaborn as sns
```

```python
#Where you did better? Clearly second one.
norm.cdf((86-75)/15,0,1)
norm.cdf((498-350)/110,0,1)
```

```python
x = np.linspace(-3,3,1000)
y = norm.pdf(x,0,1)
```

```python
plt.plot(x,y)
```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=20,metadata = dict(artist='Me'),bitrate=1800)
fig = plt.figure()
plt.xlim = (-3,3)
plt.ylim = (0,1)
# plt.xlabel('Marks')
# plt.ylabel('Probability')
# plt.title('Normal Distribution of Marks')
def animate(i):
    p = sns.lineplot(x=x[:i*50],y=y[:i*50],color="r")
ani = animation.FuncAnimation(fig, animate, frames=20, repeat=True)
plt.show()
ani.save('HeroinOverdosesJumpy.mp4', writer=writer)
```
