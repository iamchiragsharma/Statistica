```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(35,95,100)
```

```python
y  =norm.pdf(x,65,10)
l = norm.cdf(88,65,10)
l
```

```python
plt.plot(x,y,'r')
```

```python
#Normalization Technique
z = (x-65)/10
z
```

```python
y1 = norm.pdf(z,0,1)
plt.plot(z,y1,'g')
```

```python
mba = np.linspace(462,559,100)
dist = norm.pdf(mba,495,12.5)
l = norm.cdf(525,495,12.5)
print(l)
plt.plot(mba,dist)
```

```python
zscore = (mba-495)/12.5
newnorm = norm.pdf(zscore,0,1)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,1,1)
plt.plot(zscore,newnorm,'r')
plt.subplot(2,1,2)
plt.plot(z,y1,'g')
```
