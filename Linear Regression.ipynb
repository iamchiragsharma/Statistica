```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = (15,8)

#Reading Data 
data  = pd.read_csv("Datasets/headbrain.csv")
print(data.shape)
print(data.head())
```

```python
#Result is an array
X = data['Head Size(cm^3)'].values
#Result is an pandas Series
#x = data['Head Size(cm^3)']
#type(x)
Y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
```

```python
n = len(X) #Total number of values
num = 0
den = 0
for i in range(n):
    num+=(X[i]-mean_x)*(Y[i]-mean_y)
    den+=(X[i]-mean_x)**2
m = num/den
c = mean_y - (m*mean_x)
```
