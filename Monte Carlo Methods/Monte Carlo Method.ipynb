```python
import random 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12,8)
```

```python
def rolldice():
    number = random.randint(1,100)
    if number<=48:
        return "Win"
    else:
        return "Lose"
```

```python
def you(money,perBet,times):
    #gameServed keeps track of how many games you have played
    gameServed = 1

    #Luckily to prove you I asked Henry to Keep Performance Record
    Performance = {}

    while gameServed <= times:
        if rolldice() == "Win":
            sweetMoney = money + perBet
        else:
            sweetMoney = money - perBet
        Performance[gameServed] = sweetMoney
        gameServed +=1
    return Performance
        
```

```python
myamount  = you(100,10,10)
plt.plot(myamount.keys(),myamount.values())
plt.xlabel('No of games')
plt.ylabel('Your Money')
plt.savefig('5games.png')
```
