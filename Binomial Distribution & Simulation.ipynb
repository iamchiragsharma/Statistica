```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = np.arange(0,11) #If any value for k is such that k>n than the binomial probability for that event X=k
#would be zero since there is no way of getting successful events k such that k>n.
#You just cannot have 13 successful trials when no of total trials was 11
print(k)
binomial = binom.pmf(k,10,0.5)
binomial
```

```python
plt.plot(k,binomial,'o--')

plt.xlabel("No of successes")
plt.ylabel("Binomial Distribution")
plt.show()
```

```python
#Running a binomial simulation
binom_sim = binom.rvs(n=10,p=0.5,size=2000)
print(type(binom_sim))
print(np.mean(binom_sim))
plt.hist(binom_sim,bins=20,density=True)
plt.show()
```

```python
#Calculating a cummulative binomial probality distribution
binomial = binom.cdf(5,15,0.65)
print(binomial)
```
